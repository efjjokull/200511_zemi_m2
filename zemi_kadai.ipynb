```python
import pandas as pd
import numpy as np
from IPython.display import display

df = pd.read_csv("kadai.csv")

display(df.head(10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

X = df[["count_0", "count_1", "count_2", "count_3", "count_4", "count_5", "period"]]
y = df['count_pred']
```

```python
display(X.head(10))
```

```python
display(y[:10])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)

y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
print("Training set score: ", lr.score(X_train, y_train))

print("Test set score: ", lr.score(X_test, y_test))
```

```python
# RMSE と R^2による評価
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))

print('RMSE train: %.3f, test: %.3f' % (
        np.sqrt(mean_squared_error(y_train, y_train_pred)),
        np.sqrt(mean_squared_error(y_test, y_test_pred)))
     )
```

```python
significant = pd.DataFrame({"features":X.columns, "coefficient":lr.coef_})
display(significant)
```

```python
# Ridge
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=10).fit(X_train, y_train)
print("X_train score: {:.2f}".format(ridge.score(X_train, y_train)))
print("X_test score: {:.2f}".format(ridge.score(X_test, y_test)))

y_train_pred = ridge.predict(X_train)
y_test_pred = ridge.predict(X_test)
```

```python
# lasso(デフォルト)
from sklearn.linear_model import Lasso
import numpy as np

lasso = Lasso().fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))

y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)
```
